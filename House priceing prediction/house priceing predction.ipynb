```python
from sklearn.model_selection import train_test_split
data = f_data
X_train,X_valid=train_test_split(data)
y_train = X_train.drop(['SalePrice'],axis=1)
y_valid = X_valid.drop(['SalePrice'],axis=1)

```

```python
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor

model1 = XGBRegressor(n_estimators=500,learning_rate=0.01)
model1.fit(X_train, y_train)
predictions = model1.predict(X_valid)

mse = mean_squared_error(y_valid, predictions)
print(mse)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor(n_estimators=50, random_state=0)
model2.fit(X_train, y_train)
predictions = model2.predict(X_valid)

mse = mean_squared_error(y_valid, predictions)
print(mse)
```
